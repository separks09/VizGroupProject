```python
import pandas as pd
from flask import Flask, jsonify, render_template
#Creates dataframe of season stats
a=pd.read_csv('nfldata.csv')

#Creates list of teams, seasons
listofteams=[i for i in a.groupby('team').groups.keys()]
listofseasons=[i for i in a.groupby('season').groups.keys()]

#Sums data for each season for each team
finallist=[]
for team in listofteams:
    teamstats={}
    teamstats['team']=team
    for year in listofseasons:
        try:
            yearstats=dict(a.groupby('team').get_group(team).groupby('season').get_group(year).sum())
            yearstats['season']=year
            del yearstats['week']
            del yearstats['wins']
            yearstats['team']=team
            finallist.append(yearstats)
        except:
            yearstats={'season': year,
             'turnovers': 0,
             'pass_yds': 0,
             'rush_yds': 0,
             'team': team,
             'pts': 0,
             'pts_against': 0,
             'opp_pass_yd': 0,
             'opp_rush_yds': 0,
             'takeaways': 0,
             'win': 0}
            finallist.append(yearstats)
a=pd.DataFrame(finallist)
a['teamseason']=[a['team'][i]+' ' + str(a['season'][i]) for i in range(len(a['team']))]
```

```python
listofstats=[i for i in a]
del(listofstats[-3])

finaljson=[]
for team in listofteams:
    teamdictionary={}
    teamdictionary['team']=team
    for stat in listofstats:
        statlist=[]
        for index in range(len(a['season'])):
            if a['team'][index]==team:
                statlist.append([int(a['season'][index]),int(a[stat][index])])
        teamdictionary[stat]=statlist
    finaljson.append(teamdictionary)

#Exports to json file
import json

with open('data.json', 'w') as fp:
    json.dump(finaljson, fp)
```
